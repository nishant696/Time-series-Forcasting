```python
train_data['date'] = pd.to_datetime(train_data['date'])
hol['date'] = pd.to_datetime(hol['date'])
```

```python
train_data
```

```python
#holidays_events_df['date'] = pd.to_datetime(holidays_events_df['date'])
#train_df['date'] = pd.to_datetime(train_df['date'])

holiday_types = ['Holiday', 'Additional', 'Bridge', 'Work Day', 'Transfer', 'Event']
for holiday_type in holiday_types:
    hol[holiday_type] = (hol['type'] == holiday_type).astype(int)

train_data = pd.merge(train_data, hol[['date'] + holiday_types], on='date', how='left')
train_data[holiday_types] = train_data[holiday_types].fillna(0).astype(int)

train_data.head()
```

```python
holiday_sales_means = {}
for holiday_type in holiday_types:
    filter_rows = train_data[train_data[holiday_type]==1.0]
    holiday_sales_means[holiday_type] = filter_rows['sales'].mean()

no_holiday = train_data[(train_data[holiday_types].sum(axis=1)==0)]
holiday_sales_means['No_holiday'] = no_holiday['sales'].mean()
```

```python
fig,ax = plt.subplots()
ax.bar(holiday_sales_means.keys(), holiday_sales_means.values())
ax.set_xlabel('holidays')
ax.set_ylabel('total sales')
plt.xticks(rotation = 90)
plt.show()
```

```python
family_types = [col for col in train_data.columns if col.startswith('family_')]
```

sales by day of month

```python
train_data['day_of_month'] = train_data['date'].dt.day
day_of_month_sales_sum = train_data.groupby('day_of_month')['sales'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(day_of_month_sales_sum['day_of_month'], day_of_month_sales_sum['sales'], color='grey')
plt.title('Total Sales by Day of Month')
plt.xlabel('Day of Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 32)) 
plt.ylim(20000000, day_of_month_sales_sum['sales'].max() * 1.1) 
plt.tight_layout()
plt.show()
```
